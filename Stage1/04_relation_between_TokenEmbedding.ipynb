```python

import gensim.downloader as api
import numpy as np
model = api.load("word2vec-google-news-300")
```

```python

word_vectors = model
# Let us look how the vector embedding of a word look like
print(word_vectors['computer'])

#300 dimensonal vector size of every token in word2vec model
print(word_vectors['cat'].shape)

# king + Women - Man = ?
print(word_vectors.most_similar(positive = ['king', 'women'], negative = ['man']))

# Example of calculating similarity
print(word_vectors.similarity('women', 'man'))
print(word_vectors.similarity('king', 'queen'))
print(word_vectors.similarity('uncle', 'aunt'))
print(word_vectors.similarity('boy', 'girl'))
print(word_vectors.similarity('nephew', 'niece'))
print(word_vectors.similarity('paper', 'water'))
```

```python


# Finding most similar to given words
print(word_vectors.most_similar("tower", topn=4))

words = ['man', 'women', 'semiconductor', 'earthworm', 'nephew', 'niece']

# calculate the vector difference
vector_diff = []

for i in range(0, 6, 2) :
    vector_diff.append(word_vectors[words[i]] - word_vectors[words[i+1]])

print(vector_diff)

magnitude_of_diff = []
for diff in vector_diff :
    magnitude_of_diff.append(np.linalg.norm(diff))

for i_magnitude, i_word in enumerate(range(0, 6, 2)) :
    print(f"The magnitude of the difference between {words[i_word]} - {words[i_word + 1]} : {magnitude_of_diff[i_magnitude]}")
    
# Note : Less the magnitude is, means closer the vector embedding are and their meaning
```
